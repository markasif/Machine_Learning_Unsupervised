```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
data.shape
```

```python
scaler=MinMaxScaler()
data_scaled=pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
data_scaled.head()
```

# Dendrogram to decide no.of clusters

```python
#import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
plt.figure(figsize=(10,7))
plt.title("Dentrogram")
#the linkage=measure of similarity between clusters of data points
#the algorithm will merge the pairs of clusters that minimise the critertion
#method used for detemining the distance between newly formed cluster and 
#the other data points---single,complete,average etc..
#method="ward" uses the word varience minimisation algorithm
dend=dendrogram(linkage(data_scaled,method='ward'))
```

### 'single' uses the minimum distance of the between all observation of the two sets

```python
plt.figure(figsize=(10,7))
plt.title('Dentrogram')
dend=dendrogram(linkage(data_scaled,method='ward'))
plt.axhline(y=10,color='r',linestyle='--')
#axhline:- to draw a horizontal line on a plot
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,linkage='ward')
```

```python
clusters=model.fit_predict(data_scaled)
clusters
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_scaled['Fresh'],data_scaled['Detergents_Paper'],c=clusters)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('shopping-data.csv')
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Genre'] = data['Genre'].map({'Male': 0, 'Female': 1})
```

```python
scaler=MinMaxScaler()
data_scale=pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
data_scale.head()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
plt.figure(figsize=(10,7))
plt.title("dendrogram")
dend=dendrogram(linkage(data_scale,method="ward"))
```

```python
plt.figure(figsize=(10,7))
plt.title("Dendrogram")
dend=dendrogram(linkage(data_scale,method="ward"))
plt.axhline(y=8,color='green',linestyle='--')
                
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,linkage='ward')
```

```python
cluster=model.fit_predict(data_scale)
cluster
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_scale['Annual Income (k$)'],data_scale['Spending Score (1-100)'],c=cluster)
```
